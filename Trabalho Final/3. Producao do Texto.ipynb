## Componentes

```python
from statsmodels.tsa.seasonal import MSTL
```

```python
stl_kwargs = {"seasonal_deg": 0} 
model = MSTL(demand_series, periods=(24, 24 * 7), stl_kwargs=stl_kwargs)
res = model.fit()
```

```python
res.plot();
```

```python
seasonal = res.seasonal # contains both seasonal components
trend = res.trend
residual = res.resid
```

```python
trend.plot(figsize=(15,3))
```

```python
seasonal.iloc[:1000].plot(figsize=(15,3))
```

```python
residual.plot(figsize=(15,3))
```

```python
demand_series_transform1 = (demand_series - trend) 
demand_series_transform1.iloc[:5000].plot(figsize=(15,3))
```

```python
seasonal.columns
```

```python
demand_series_transform2 = (demand_series_transform1 - seasonal.iloc[:,0]) 
demand_series_transform2.iloc[:5000].plot(figsize=(15,3))
```

```python
demand_series_transform3 = (demand_series_transform2 - seasonal.iloc[:,1]) 
demand_series_transform3.iloc[:5000].plot(figsize=(15,3))
```

```python
(demand_series).iloc[window_start:window_end].rename("demanda real").plot(style='--', color='gray', figsize=(15,3), legend=True)
(trend + seasonal.iloc[:,0]).iloc[window_start:window_end].rename("tendência+diário").plot(figsize=(15,3), legend=True, color='tab:orange')
(trend + seasonal.iloc[:,0] + seasonal.iloc[:,1]).iloc[window_start:window_end].rename("tendência+semanal+diário").plot(figsize=(15,3), legend=True, color='tab:blue')
(trend).iloc[window_start:window_end].rename("tendência").plot(figsize=(15,3), legend=True, color='tab:green')

plt.legend(loc="upper right", prop={'size': 9})
plt.savefig("Imagens/decomposicao_mstl.png")
```

## Regressão Linear para Obter Pesos

```python
from sklearn.linear_model import LinearRegression
```

```python
input_data = df.set_index("datehour").iloc[window_start:window_end:,11:]
```

```python
output_data = df.set_index("datehour").iloc[window_start:window_end:,0]
```

```python
output_data
```

```python
modelo = LinearRegression(fit_intercept=False).fit(input_data, output_data)
```

```python
# modelo.coef_
```

```python
# maiores e menores
sel_indexes = pd.Series(modelo.coef_).nlargest(5).index.tolist() + pd.Series(modelo.coef_).nsmallest(5).index.tolist()
```

```python
(input_data * modelo.coef_).iloc[:24*7,2:30].plot(figsize=(7,3), legend=False)

plt.savefig("Imagens/weighted_fourier_components.png")
```

```python
output_data.iloc[:24*7].rename("demanda real").plot(style='--', color='tab:orange', figsize=(7,3), legend=True)
(input_data @ modelo.coef_).iloc[:24*7].rename("predicao").plot(color='tab:blue', figsize=(7,3), legend=True)

plt.legend(loc="lower left", prop={'size': 9})
plt.savefig("Imagens/fourier_component_prediction.png")
```
