## Tratamento dos Dados

```python
def get_data(df, column_x, column_y):
    X = df[column_x]
    Y = df[column_y]
    return X, Y
```

# Classificação com Regressão Logística

Para a regressão logística é importante que façamos a normalização dos dados para a convergência.

Iniciaremos com uma tentativa simples usando todas as features contra a coluna de labels.

```python
# separacao de treinamento e validacao
df_data_t1_train, df_data_t1_test = train_test_split(df_data_t1, train_size=0.7)
```

```python
# Obtendo os dados de treinamento
cols = df_data_t1_train.columns.tolist()
cols.remove("diagnosis")

X, Y = get_data(df_data_t1_train, cols, "diagnosis")
```

```python
# Para chegar a convergência precisamos escalar
scaler = StandardScaler().fit(X)
X = scaler.transform(X)
Y = np.array(Y)
```

```python
# Treinamento do Modelo
model = LogisticRegression().fit(X,Y)
```

```python
# Predizendo a partir do modelo e os dados de teste
X_test, Y_test = get_data(df_data_t1_test, cols, "diagnosis")

# Para chegar a convergência precisamos escalar
X_test = scaler.transform(X_test)
Y_test = np.array(Y_test)

Y_pred = model.predict(X_test) 
```

```python
from sklearn import metrics

print("Acurácia:", metrics.accuracy_score(Y_test.tolist(), Y_pred))
print("Precision:", metrics.precision_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))
print("Recall:", metrics.recall_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))
print("F1-score:", metrics.f1_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))
```

```python
matriz_confusao = metrics.confusion_matrix(Y_test.tolist(), Y_pred, labels=model.classes_)
metrics.ConfusionMatrixDisplay(confusion_matrix=matriz_confusao, display_labels=model.classes_).plot()
```

Tivemos um excelente resultado, com todas as seções da Matriz de Confusão com excelentes valores. 

Tivemos praticamente todas as respostas corretas e quase todas as incorretas perto de zero.

## Modelo Pronto

```python
from sklearn import metrics
```

```python
def regressao_logistica(df_data_t1):
    # separacao de treinamento e validacao
    df_data_t1_train, df_data_t1_test = train_test_split(df_data_t1, train_size=0.7, random_state=42)

    # Obtendo os dados de treinamento
    cols = df_data_t1_train.columns.tolist()
    cols.remove("diagnosis")

    X, Y = get_data(df_data_t1_train, cols, "diagnosis")

    # Para chegar a convergência precisamos escalar
    scaler = StandardScaler().fit(X)
    X = scaler.transform(X)
    Y = np.array(Y)

    # Treinamento do Modelo
    model = LogisticRegression().fit(X,Y)

    # Predizendo a partir do modelo e os dados de teste
    X_test, Y_test = get_data(df_data_t1_test, cols, "diagnosis")

    # Para chegar a convergência precisamos escalar
    X_test = scaler.transform(X_test)
    Y_test = np.array(Y_test)

    Y_pred = model.predict(X_test) 

    print("Acurácia:", metrics.accuracy_score(Y_test.tolist(), Y_pred))
    print("Precision:", metrics.precision_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))
    print("Recall:", metrics.recall_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))
    print("F1-score:", metrics.f1_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))

    matriz_confusao = metrics.confusion_matrix(Y_test.tolist(), Y_pred, labels=model.classes_)
    metrics.ConfusionMatrixDisplay(confusion_matrix=matriz_confusao, display_labels=model.classes_).plot()
```

# Seleção de Features

## Por variância dos valores

Uma forma de selecionar features é através de sua variância/desvio padrão. Se os dados forem muito constantes eles não serão relevantes para escolhermos as classes.

```python
cols = df_data_t1.columns.tolist()
cols.remove("diagnosis")

colunas_selecionadas = df_data_t1[cols].std().sort_values(ascending=False)[:20].index.tolist()
colunas_selecionadas
```

```python
colunas_selecionadas.append("diagnosis")
dados_selecionados = df_data_t1.loc[:, colunas_selecionadas]

regressao_logistica(dados_selecionados)
```

Em geral esse método piorou nossos resultados anteriores, menos o F1. Provavelmente estamos removendo colunas importantes.

## Utilizando apenas as médias

Removeremos as informações que não são médias para deixar apenas uma feature por tipo. 

```python
colunas_selecionadas = df_data_t1.columns[df_data_t1.columns.str.endswith("_mean")].tolist()
colunas_selecionadas
```

```python
colunas_selecionadas.append("diagnosis")
dados_selecionados = df_data_t1.loc[:, colunas_selecionadas]

regressao_logistica(dados_selecionados)
```

Esse caso tivemos resultados ainda piores, provavelmente a média é uma medida muito geral.

# Balanceamento dos Dados

Balanceamento de classes costuma ser um problema relevante para a generalização de classes pequenas e pudemos ver que os dados estão desbalanceados.

Utilizarei o oversampler aleatório no dataset de treinamento, deixando o de teste desbalanceado proporcionalmente. O oversampler escolhe aleatoreamente uma linha para replicar.

```python
df_data_t1.head()
```

```python
# separacao de treinamento e validacao
df_data_t1_train, df_data_t1_test = train_test_split(df_data_t1, train_size=0.7, random_state=42)
```

```python
# Obtendo os dados de treinamento
cols = df_data_t1_train.columns.tolist()
cols.remove("diagnosis")

X, Y = get_data(df_data_t1_train, cols, "diagnosis")
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
X_balanceado, Y_balanceado = ros.fit_resample(X, Y)
```

```python
# Classes balanceadas
pd.Series(Y_balanceado).value_counts().plot.bar()
```

```python
# Para chegar a convergência precisamos escalar
scaler = StandardScaler().fit(X_balanceado)
X_balanceado = scaler.transform(X_balanceado)
Y_balanceado = np.array(Y_balanceado)

# Treinamento do Modelo
model = LogisticRegression().fit(X_balanceado,Y_balanceado)
```

```python
# Predizendo a partir do modelo e os dados de teste
X_test, Y_test = get_data(df_data_t1_test, cols, "diagnosis")

# Para chegar a convergência precisamos escalar
X_test = scaler.transform(X_test)
Y_test = np.array(Y_test)

Y_pred = model.predict(X_test) 
```

```python
print("Acurácia:", metrics.accuracy_score(Y_test.tolist(), Y_pred))
print("Precision:", metrics.precision_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))
print("Recall:", metrics.recall_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))
print("F1-score:", metrics.f1_score(Y_test.tolist(), Y_pred, average='binary', pos_label='M'))

matriz_confusao = metrics.confusion_matrix(Y_test.tolist(), Y_pred, labels=model.classes_)
metrics.ConfusionMatrixDisplay(confusion_matrix=matriz_confusao, display_labels=model.classes_).plot()
```

De forma geral conseguimos melhorar nossos resultados iniciais, piorando apenas a precisão.
